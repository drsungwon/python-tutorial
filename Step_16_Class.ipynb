## 프로그래머 정의 클래스 방법 이해하기

클래스의 가장 중요한 개념은 서로 상관성이 큰 데이타(이후 attribute)와 함수(이후 methods)를 한데 묶는 행위입니다.<br>
물론 이런 attribute와 methods를 추출해 내는 것이 추상적인 세상에서 구체적인 것을 끌어내는 일이니 가장 어렵고 중요하지만,<br>
일단 프로그래밍 언어를 배우는 현재에서는 관련된 attribute와 methods를 묶는 작업부터 하며, 이 기능의 명령어가 ``class`` 입니다.<br>

아래의 강의노트를 토대로 학생을 클래스로 만든 예제 입니다.<br>
가장 기본적으로 학생은 유일무이하게 부여되는 독특한(unique) 학번이 있으며 id라고 부르겠으며, 학생의 이름인 name이 있습니다.<br>
이런 데이터 만으로 클래스를 시작해 봅시다.<br>

```python
# Attribute only Student class

class Student:
    id = 0
    name = ''
```

끝입니다. ``^____^`` 이제 우리는 학생을 나타내는 클래스의 기본을 만든 겁니다.<br>

그러면 클래스는 설계도이니, 실제 attribute를 저장할 객체를 만들어 보도록 합시다.<br>
아래의 코드는 실제로 Student 클래스를 기반으로, student1과 student2 객체를 만들고, 각각의 객체에 id와 name의 attribute를 부여한 예입니다.<br>

```python
student1 = Student()
student2 = Student()

student1.id = 20190001
student1.name = "Harry Potter"

student2.id = 20190002
student2.name = "Hermione Granger"
```

객체지향 프로그래밍 언어들은 일종의 암묵적인 규칙이 있습니다.<br>
즉 데이터인 attribute를 읽고 쓸때 위처럼 직접 대입을 하지 않고, 함수인 methods를 만들어서 읽고/쓰는 것입니다.<br>
그러면 바로 만들어야 할 함수가 생겼으니, Student 클래스를 개선하겠습니다.<br>

Methods들이 추가된 것이 바로 보입니다.<br>
그런데 특이한 입력 파라메타들이 모두 맨처음에 있는데, 바로 ``self`` 입니다.<br>
설계도인 클래스에서 수많은 객체들이 만들어 지는데, 정작 methods는 클래스안에 하나만 존재합니다.<br>
즉 객체들마다 데이타인 attribute는 위의 예제처럼 학생 하나 하나 마다 다르지만, 사실상 함수인 methods는 동일한 셈입니다. <br>
학생별로 데이터는 다르지만 기능인 함수는 다를바가 없는 것이죠.<br>
따라서 클래스의 함수는 "내가 지금 어떤 객체의 데이터를 다뤄야 하는 거지?"라는 의문에 답으로 객체에 대한 정보를 갖는 ``self`` 변수를 첫번째 입력 파라메타로 받아야 합니다.<br>
그런데, 바로 이어질 예제에서 보겠지만, ``self`` 변수는 클래스의 멤버 methods 호출시 Python이 알아서 붙여주므로, 신경쓰지 않아도 됩니다.<br>

말이 복잡하면 그냥 쉽게 생각합니다 - 클래스의 멤버 함수는 맨처음 ``self`` 변수를 입력 파라메타로 쓰면 됩니다.<br>
그리고 멤버 데이터(attribute는)를 멤버 methods에서 활용할때는 ``self.``을 붙이면 됩니다.<br>

```python
# Student class with methods

class Student:
    id = 0
    name = ''
    
    def setId(self, givenID):
        self.id = givenID
        
    def getId(self):
        return self.id
        
    def setName(self, givenName):
        self.name = givenName
        
    def getName(self):
        return self.name
```

이에 위의 개선된 함수를 통해서 멤버 attribute를 읽고 쓰게 만든 예제가 다음과 같습니다.<br>

```python
student1 = Student()
student2 = Student()

student1.setId(20190001)
student1.setName("Harry Potter")
print(student1.getId())
print(student1.getName())
      
student2.setId(20190002)
student2.setName("Hermione Granger")
print(student2.getId())
print(student2.getName())
```

아래의 입력창에 위의 Student 클래스 프로그램과 바로 위의 사용 예제를 입력하고 실행해 보도록 합니다.

## 생성자(Constructor, ``__init__()``) 이해하기

위의 코드를 보다보면, 이런 생각이 들수 있습니다.<br>
Student 객체를 생성할때, 번거럽게 멤버 attribute 별로 함수를 호출해야 하나? <br>
멤버 attribute가 늘어나면 늘어날수록 호출해야 하는 함수도 늘어나는 걸까?<br>

이런 질문에 대한 답으로, 객체가 만들어 질때 멤버 attribute를 초기화 할 것이 있으면 자동으로 하고,<br>
심지어 객체가 만들어 지면 반드시 호툴해야 하는 함수(methods)도 그냥 Python 언어가 자동으로 호출해 주면 편할텐데...<br>
이런 기능을 해주는 함수가 생성자 혹은 Constructor 혹은 (Python 언어의 경우) ``__init__()`` 함수 입니다.<br>
이 함수는 객체가 만들어 질때 Python 언어가 자동으로 호출하게 됩니다.<br>

우리는 생성자를 2단계로 설명합니다.<br>
먼저 모든 객체에 대해서 동일하게 자동으로 이루어 져야 할일이 있다고 생각해 봅시다.<br>
우리는 이미 학생의 ID와 이름을 만들고, 0과 ""으로 초기화 한 것을 보았습니다.<br>
이렇게 객체가 만들어지면 모든 객체들에 대해서 가장 먼저 해야할 일을 생성자를 통해서 하는 것으로 바꾸면 다음과 같습니다.<br>
즉 생성자 함수인 ``__init__()``안에서 멤버 attribute를 정의하고, 초기화 하였습니다.<br>

```python
# Student class with methods

class Student:
    def __init__(self):
        self.id = 0
        self.name = ''
    
    def setId(self, givenID):
        self.id = givenID
        
    def getId(self):
        return self.id
        
    def setName(self, givenName):
        self.name = givenName
        
    def getName(self):
        return self.name
```

생성자가 요긴하게 쓰이는 두번째는 모든 객체에 대해서 동일한 값이 아닌 상태로 객체를 만드는 경우입니다.<br>
우리의 Student의 경우 ID와 이름이 이에 해당합니다.<br>
이런 경우 생성자에 입력 파라메타를 줄수 있도록, 생성자 함수를 다시 개선해 봅니다.<br>
아래의 코드를 보면, 생성자가 Python 언어가 자동으로 전달하는 self외에 두개의 입력 파라메타를 받는 것을 볼수 있습니다.<br>
그리고 받은 값은 ID와 이름으로서, 멤버 attribute에 초기화 해주는 것을 볼수 있습니다.<br>
이렇게 함으로서 우리는 객체를 생성할때, 객체마다 다르게 가져야 하는 값을 생성자의 파라메타로 전달할 수 있습니다.<br>

```python
# Student class with constructor

class Student:
    def __init__(self, givenID, givenName):
        self.id = givenID
        self.name = givenName
        
    def setId(self, givenID):
        self.id = givenID
        
    def getId(self):
        return self.id
        
    def setName(self, givenName):
        self.name = givenName
        
    def getName(self):
        return self.name
```
        
위의 생성자를 토대로 객체를 생성한다면, 앞서의 경우보다 단촐해 집니다.<br>

```python
student1 = Student(20190001, "Harry Potter")
student2 = Student(20190002, "Hermione Granger")

print(student1.getId())
print(student1.getName())
      
print(student2.getId())
print(student2.getName())  
```

## ``__str__()`` 적용 이해하기

위의 사용방법을 보면, 이제 객체를 생성하고 초기화 값을 주는 부분은 나름 한줄로 되어 괜찮은데, 굳이 멤버 attribute를 한줄 한줄 작성해서 봐야 하나 싶습니다.<br>
멤버 attribute의 내용을 보려면, attribute 갯수 만큼 print()를 한다는 것이 사실 정상적이지는 않으니 뭔가 방법이 있겠지요?<br>
앞서 우리는 리스트와 같이 복잡한 데이타도 print() 함수 한번으로 출력할 수 있었습니다.<br>
여러분이 직접 만드는 클래스도 이렇게 하는 것이 가능하며, 방법은 매우 쉽게 구현할 수 있습니다.<br>
즉 아래와 같이 프로그래머가 만든 클래스의 멤버 methods로 ``__str__()`` 함수를 포함하고, 이 함수에서 클래스의 attribute를 한번에 출력하게 하면 됩니다.<br>

```python
# Student class with constructor

class Student:
    def __init__(self, givenID, givenName):
        self.id = givenID
        self.name = givenName
        
    def setId(self, givenID):
        self.id = givenID
        
    def getId(self):
        return self.id
        
    def setName(self, givenName):
        self.name = givenName
        
    def getName(self):
        return self.name
                
    def __str__(self):
        msg = "id:{}, name:{}".format(self.id, self.name)
        return msg
```

그러면 호출을 해볼까요?<br>
위의 ``__str__()`` 멤버 methods를 만들었기에, 앞서의 경우보다 멤버 attribute의 출력 구문이 단촐해 집니다.<br>

```python
student1 = Student(20190001, "Harry Potter")
student2 = Student(20190002, "Hermione Granger")

print(student1.getId())
print(student1.getName())
      
print(student2.getId())
print(student2.getName())  
```

아래의 입력창에 최종 수정한 Student 클래스와 위의 객체 생성 예제를 입력하고, 직접 실행해 보기 바랍니다.

## 클래스 변수와 객체 (혹은 인스턴스) 변수 이해하기

Student 클래스의 최종 버전 기준, 클래스가 가져야 하는 attribute는 생성자 안에 있으며, ``self.``로 시작하고 있습니다.<br>
이렇게 표현되는 변수는 클래스로 부터 만들어 지는 특정 객체 들에 엮입니다.<br>
즉 "Harry Potter" 객체에 name이 "Harry Potter"로 정해지며, "Hermione Granger" 객체의 name에는 "Hermione Granger"이 저장되는 것이지요.<br>
그런데 종종 클래스 차원에서 attribute가 필요한 경우가 있는데, 대표적인 예제게 **클래스에서 만들어지는 객체의 갯수 세기**의 목적입니다.<br>
이런 경우에 쓰는 변수가 클래스에 종속된 **클래스 변수**입니다.<br>

아래의 코드는 Student 클래스에 추가적으로 클래스 변수인 countStudent를 추가한 예제입니다.<br>

```python
class Student:
    # Class variables
    countStudent = 0
    
    def __init__(self, givenID, givenName):
        # Instance (or Object) variables
        self.id = givenID
        self.name = givenName
        Student.countStudent = Student.countStudent + 1
        
    def setId(self, givenID):
        self.id = givenID
        
    def getId(self):
        return self.id
        
    def setName(self, givenName):
        self.name = givenName
        
    def getName(self):
        return self.name
                
    def __str__(self):
        msg = "id:{}, name:{}".format(self.id, self.name)
        return msg
    
    def getNumOfStudent():
        return Student.countStudent
```

countStudent 변수를 정의한 위치는 멤버 methods 들의 바깥입니다.<br>
이렇게 객체에 종속되지 않고, 클래스에 연결된 변수의 선언은 멤버 methods나 생성자가 아닌 클래스의 시작 부분에 기술합니다.<br>
사용 방법도 차이가 있습니다.<br>
생성자안을 보면, 객체가 만들어질때 마다 countStudent 변수 값을 증가시키는데, 객체에 속하지 않았기에 ``self.``로 시작하지 않고,<br>
``OOO 클래스에 속한``이라는 의미로 클래스의 이름은 ``Student.``로 시작하는 것을 볼수 있습니다.<br>
마찬가지로 이 생성된 학생수를 돌려주는 getNumOfStudent() 함수에서도 ``self`` 입력 파라메타가 없습니다.<br>

자, 그러면 실제로 아래의 코드를 통해서 실행을 해봅니다.<br>

```python
print(Student.getNumOfStudent())

student1 = Student(20190001, "Harry Potter")
student2 = Student(20190002, "Hermione Granger")

print(student1)
print(student2)  

print(Student.getNumOfStudent())
```

첫줄은 객체를 만들지 않은 상황에서 클래스의 countStudent 변수를 출력하기에 현재 초기화된 0을 출력합니다.<br>
그리고 마지막 줄은 당연히 두개의 객체를 만들었으므로, 2를 출력합니다.<br>
이렇게 우리는 클래스 차원에서의 attribute를 관리할 수 있습니다.<br>

위의 Student 클래스와 사용 예제 코드를 아래 입력창에서 실제 수행하여 보도록 합니다.<br>

## Private Attribute 이해하기

앞서 멤버 attribute를 직접 읽고 쓰게 하지 말고 함수(methods)를 통해서 읽고 쓰는 것이 일반적이라고 했습니다.
아예 강제적으로 attribute를 직접 읽고 쓰는 것을 막을수 있으니, 단순하게 멤버 attribute의 이름 앞에 ``__``를 하면 됩니다.
이미 우리가 만든 Student는 모든 변수가 직접 읽고 쓰지 않도록 하고 있으니, 모든 변수에 적용해 봅니다.

```python
class Student:
    # Class variables
    __countStudent = 0
    
    def __init__(self, givenID, givenName):
        # Instance (or Object) variables
        self.__id = givenID
        self.__name = givenName
        Student.__countStudent = Student.__countStudent + 1
        
    def setId(self, givenID):
        self.__id = givenID
        
    def getId(self):
        return self.__id
        
    def setName(self, givenName):
        self.__name = givenName
        
    def getName(self):
        return self.__name
                
    def __str__(self):
        msg = "id:{}, name:{}".format(self.__id, self.__name)
        return msg
    
    def getNumOfStudent():
        return Student.__countStudent
```

그리고 나서 다음의 줄을 실행하면, 에러가 납니다. 강제적으로 클래스/객체의 멤버 attiribute를 직접 읽고 쓰는 것을 막은 것입니다.

```python
print(Student.__countStudent)
```

## 상속 (Inheritance) 이해하기

이름이 상속이 되니 마치 생물학을 이야기 하는 것 같은데, 얼추 맞습니다.<br>
이제 우리는 프로그래머가 만든 클래스에 필요한 경우 추가적인 attribute와 methods를 집어 넣어서 새로운 클래스를 만들수 있습니다.<br>
왜 이렇게 할까요? 현대 프로그래머에게 가장 중요한 무기는 생산성 입니다.<br>
즉 정해진 시간에 이미 남이 만들어 놓은 좋은 코드들을 충분하게 ``재사용``해서, 빠르고 안전한 코드를 조기에 확보하여 남보다 빠르게 서비스를 하는 겁니다.<br>
따라서 다른 사람 혹은 내가 만든 과거의 코드에 필요한 기능을 추가만해서, 이미 만든 코드가 안전하게 돌아가기에 새로운 기능에만 집중하는 겁니다.<br>
Student 클래스를 이제 구분하여 하나는 "재학생", 그리고 하나는 "졸업생"으로 확장해 보도록 합니다.<br>
우리는 졸업생을 만들어 봅시다.<br>

```python
class GraduatedStudent(Student):
    
    def __init__(self, givenId, givenName, givenYear):
        self.__graduatedYear = givenYear
        super().__init__(givenId, givenName)
        
    def __str__(self):
        msg = super().__str__() + ", graduation:{}".format(self.__graduatedYear)
        return msg
```

첫째로 클래스를 정의하는 구문에 GraduatedStudent(Student)처럼 괄호를 열고, 앞서 이미 만들어져서 이제 GraduatedStudent 클래스가 개선할 클래스의 이름을 넣습니다.<br>
통상 이렇게 확장의 대상이 되는 클래스를 부모 혹은 베이스(base) 클래스라고 합니다.<br>
GraduatedStudent 클래스 처럼 base 클래스에서 확장된 클래스는 자식 혹은 유전된(derived) 클래스라고 합니다.<br>
Derived 클래스에는 base 클래스의 모든 attribute와 methods가 고스란히 포함되어 있습니다.<br>

둘째로 객체 attribute로 ``__graduatedYear``이 추가된 것을 볼수 있습니다.<br>
``__``는 private 형태로, 직접 변수에 읽고 쓰지 못하게 하고, 멤버 methods를 통해서 읽고 쓰도록 막은 것입니다.<br>

셋째로 derived 클래스도 생성자를 갖는 것을 볼수 있습니다.<br>
그런데 derived 클래스에 base 클래스의 attribute가 모두 포함되기에, 졸업생의 경우도 ID와 Name을 받아야 하지만,<br>
이 정보는 Student 클래스 안에서 관리되므로, GraduatedStudent는 본인 보다 조상인 base 클래스 정보에 접근하기 위하여 super() 함수를 사용합니다.<br>
``super().__init__(givenId, givenName)``의 의미는 GraduatedStudent의 조상인 base 클래스의 생성자를 호출하라는 의미입니다.<br>
따라서, GraduatedStudent의 생성자는 본인이 관리하는 졸업 일자를 초기화하고, 바로 이름과 ID는 base 클래스의 객체에 전달하는 것입니다.<br>

넷째로 derived 클래스는 base 클래스와 같은 이름의 함수를 가질수 있는데, 이 경우 derived 클래스의 함수가 호출이 되면서,<br>
필요한 경우는 base 클래스의 함수를 부를수 있으나, 필요 없으면 derived 클래스의 함수 작업 만으로 종료합니다.<br>
이렇게 derived 클래스가 base 클래스와 똑같은 이름의 함수를 갖는 것을 오버라이딩 이라고 합니다.<br>
우리는 derived 클래스의 ``__str__()`` 함수 호출에서 이를 볼수 있는데, 앞서 배운 super() 함수를 써서,<br>
문자열로 출력하기 용이한 형태로 만들어 주는 ``__str__()`` 함수에 대해서 졸업 연도의 정보는 GraduatedStudent 클래스 함수에서 채우지만,<br>
ID와 이름에 대한 정보로 문자열을 만드는 것은 base 클래스의 ``__str__()`` 함수를 통해서 획득하는 것을 볼수 있습니다.<br>

아래의 입력창에 GraduatedStudent 클래스를 입력하고, 다음의 두줄을 함께 실행해서 동작을 이해합니다.<br>

```python
student1 = GraduatedStudent(20190001, "Harry Potter", 2023)

print(student1)
```
